```python
# find the all valid association rules with apriori algo
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules 
df = pd.read_excel('Ass2.xlsx', sheet_name='Sheet1')

# split the data into individual items
df['Item'] = df['Item'].apply(lambda x: x.split(','))

# hot encoding df['Item']
df_hot_encoded = df.join(df['Item'].str.join('|').str.get_dummies())
df_hot_encoded = df_hot_encoded.drop('Item', axis=1)

# find the freq item
freq_itemsets = apriori(df_hot_encoded.drop('Transection_ID', axis=1), min_support=0.6, use_colnames=True)

# find the association rules
rules = association_rules(freq_itemsets, metric='confidence', min_threshold=0.75)
rules = rules.sort_values(by='lift', ascending=False)
rules
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
# TF-IDF matrix of Q19_text.txt
# Read the text file
with open('Q19_text.txt', 'r') as file:
    text = file.read()

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Compute the TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform([text])

# Convert the TF-IDF matrix to a DataFrame
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Normalize the TF-IDF matrix
normalized_tfidf = df_tfidf.div(df_tfidf.sum(axis=1), axis=0)

normalized_tfidf
```

```python

# Calculate the TF-IDF scores
tfidf_scores = normalized_tfidf.iloc[0]

# Sort the TF-IDF scores in descending order
sorted_tfidf_scores = tfidf_scores.sort_values(ascending=False)

# Get the top three tokens that are most unique to Q19_text.txt
top_three_tokens = sorted_tfidf_scores.head(3)

top_three_tokens
```
