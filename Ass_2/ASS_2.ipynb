```python
# find the all valid association rules with apriori algo
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules 
df = pd.read_excel('Ass2.xlsx', sheet_name='Sheet1')

# split the data into individual items
df['Item'] = df['Item'].apply(lambda x: x.split(','))

# hot encoding df['Item']
df_hot_encoded = df.join(df['Item'].str.join('|').str.get_dummies())
df_hot_encoded = df_hot_encoded.drop('Item', axis=1)

# find the freq item
freq_itemsets = apriori(df_hot_encoded.drop('Transection_ID', axis=1), min_support=0.6, use_colnames=True)

# find the association rules
rules = association_rules(freq_itemsets, metric='confidence', min_threshold=0.75)
rules = rules.sort_values(by='lift', ascending=False)

rules
```

```python
# clean column anteceent and consequents, delete the brackets
rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(list(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(list(x)))
rules
```

```python
# export the rules to excel
rules.to_excel('Ass2_rules.xlsx', sheet_name='Sheet1', index=True)
```

```python
# remove stop words and punctuation, then compute the TF-IDF matrix
from sklearn.feature_extraction.text import TfidfTransformer, TfidfVectorizer
import pandas as pd
import string
from nltk.corpus import stopwords

# Read the text file
with open('Q19_text.txt', 'r') as file:
    text = file.read()
    
# # Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# # Remove stop words
stop_words = set(stopwords.words('english'))
text = ' '.join([word for word in text.split() if word.lower() not in stop_words])

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Compute the TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform([text])

# Convert the TF-IDF matrix to a DataFrame
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Normalize the TF-IDF matrix with Tf-idf transformer
normalized_tfidf = TfidfTransformer(norm='l1').fit_transform(df_tfidf).toarray()

# map the normalized tf-idf to the words
normalized_tfidf = pd.DataFrame(normalized_tfidf, columns=vectorizer.get_feature_names_out())
normalized_tfidf
```

```python
# read Q19_text_2.txt as doc2
with open('Q19_text_2.txt', 'r') as file:
    doc2 = file.read()
    
# read Q19_text.txt as doc1
with open('Q19_text.txt', 'r') as file:
    doc1 = file.read()

# remove punctuation and stop words and convert to lower case for doc1 and doc2
doc1 = doc1.translate(str.maketrans('', '', string.punctuation))
doc1 = ' '.join([word for word in doc1.split() if word.lower() not in stop_words])
doc1 = doc1.lower()

doc2 = doc2.translate(str.maketrans('', '', string.punctuation))
doc2 = ' '.join([word for word in doc2.split() if word.lower() not in stop_words])
doc2 = doc2.lower()

# map the frequency of each word in doc1 and doc2 in the tokens, and convert to dataframe
doc1_freq = pd.Series(doc1.split()).value_counts()
doc2_freq = pd.Series(doc2.split()).value_counts()

# convert the frequency to dataframe
doc1_freq = pd.DataFrame(doc1_freq).T
doc2_freq = pd.DataFrame(doc2_freq).T

doc1_freq
```

```python
# merge the two dataframes
df_freq = pd.concat([doc1_freq, doc2_freq], ignore_index=True)
df_freq = df_freq.fillna(0)
df_freq
```
